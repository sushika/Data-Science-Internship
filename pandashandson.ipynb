```python
import pandas as pd
#creating a dataframe
data = {
    'cars':['santro' , 'BMW' , 'Audi' , 'Jeep'],
    'duration':[50,49,45,60]
}
print(pd.DataFrame(data))# a dataframe is a table of data
```

```python
#creating pandas series using series() from a list of values
import pandas as pd
fruits = ['apple','banana','mango']#list of data (fruits)
myd = pd.Series(fruits)#series is like a coloumn in  a Dataframe
print(myd)
```

```python
#lABLES in pandas
import pandas as pd
#if no lables are mentioned the first value has 0 as its lable and the successive value has 1 
data = {
    'cars':['santro' , 'BMW' , 'Audi' , 'Jeep'],
    'duration':[50,49,45,60]
}
print(pd.DataFrame(data))
#assigning lables manually using index
print('assigning lables manually to the data')
fruits = ['apple','banana','mango']#list of data (fruits)
myd = pd.Series(fruits,index = ['a','b','c'])
print(myd)
#series can also be created using Dictionaries or key objects
schedule = {'monday':'maths','tuesday':'science','wednesday':'socialstudies','thursday':'english','friday':'PET'}
print(pd.Series(schedule))




```

```python
import pandas as pd
#locating specific rows from the dataset
data = {
    'Days':['mondays','tuesday','wednesday','Thursday','friday'],
    'subjects':['english','science','maths','social','physicaledu'],
    'faculty':['Ram','Ajay','Radha','Anil','Abhay']
}
df = pd.DataFrame(data)
print(df.loc[3])
#outputs the data in 3rd index of the dataframe
print(df.loc[[0,3]])#gives the data from the 0th and 3rd indices
```

```python
#removing the empty cells from the dataset
df.dropna()
#removes the emptycells from the dataset and displays a new dataframe with no empty cells however if inplace is set to True the actual dataset is manipulated
```

```python
x=df['city-mpg'].mean()
print(df['city-mpg'].fillna(x))#emptycells are replaced with the mean of the attribute values
#similarly they can be replaced with the median values too
y = df['normalized-losses'].median()
print(df['normalized-losses'].fillna(y))
#replacing with mode i.e most frequently occuring values
z= df['bore'].mode()
print(df['bore'].fillna(z))
```

```python
#finding duplicates if any in the dataset
import pandas as pd
df = pd.read_csv('data.csv')
print(df.duplicated().to_string())
x = df.drop_duplicates()#removes the duplicated rows from the dataset
print(x.to_string())
```
