```python
import seaborn as sns
import matplotlib.pyplot as plt
  
# loading dataset
data = sns.load_dataset("mpg")
  
# draw regressionplot
sns.lmplot(x = "horsepower", y = "mpg",data = data,hue ='origin',markers = ['*','v','o'])
#markers are used to represent various elements in the hue attribute
  
# show the plot
plt.show()
```

```python
'''in the above graph a linear regression is plot against mpg(a dependent variable) and horsepower
of vehiches from various origins like usa,japan and europe.
'''
```

```python
import seaborn as sns
df = sns.load_dataset('iris')

sns.boxplot( x=df["species"], y=df["sepal_length"] )
'''the above given graph shows the box plot considering minimum ,fist quartile,median , third quartile, and the maximum sepal lengths of the 3 species . It is observed that the maximum sepallength of the species setosa is comparitively lesser than the other two species versicolor and virginica 
where sepal length of verginica having the longest sepal length . '''
```

```python
sns.violinplot(x='species',y='sepal_length',data=df)
'''A violin plot is as similar as boxplot . it is mostly used to show multiple data in several units'''
```

```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('automobile.csv')
print(df.head())
sns.histplot(x = df['num-of-doors'],y= df['body-style'],data = df)
```

```python
'''The above histogram plot gives information about the body style of various cars and number of doors they comprise. 
it is evident that hatchback body-style cars and sedan body-style cars have both two and four doors  ,
hardtop body-style and convertible body-style cars only have 2 doors whereas wagon body-styled cars have 4 doors  '''
```

```python
 sns.stripplot(x="species",y="petal_width", data=df)
    #stripplot in seaborn is used to draw the scatter plot with a specific category
```

```python
'''A scatter plot drawn against species as a categorical variable and the peta_width of the species'''
```

```python
import matplotlib.pyplot as plt
import numpy as np
#assigning x axis elements and y axis elements
xpoints = df['petal_length']
ypoints = df['petal_width']
plt.plot(ypoints,linestyle= '--')
#labling x axis and y axis elements
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.grid()
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
  
data_set = np.random.rand( 10 , 10 )
ax = sns.heatmap( data_set , linewidth = 0.5 , cmap = 'Blues' )
  
plt.title( "Heat-map" )
plt.show()
'''A 2D heat-map is drawn considering a random matrix of 10 rows and coloumns with a colormap '''
```

```python
'''A heatmap contains values representing various shades of the same colour for each value to be plotted.
Usually the darker shades of the chart represent higher values than the lighter shade'''
```
