```python
''' Seaborn is used for visualizing the data using Python programming language.Seaborn library is built on top of Matplotlib.
It tries to make a well-defined set of hard things easy too. One of the Seaborns major features is Visualizing univariate and bivariate data , Fitting in and visualizing linear regression models , Plotting statistical time series data
Seaborn works well with NumPy and Pandas data structures . It comes with built in themes for styling Matplotlib graphics'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")#load the data set or prepare the data for visualization
```

```python
print(iris.info())
```

```python
sns.lmplot(x="sepal_width",y="sepal_length",hue="species", data=iris)#creates a linear model plot.
```

```python
 sns.barplot(x="species", y="petal_length",data=iris)
#interprets the relation between a numeric variable petal_length and a categoric variable species
```

```python
sns.boxplot(x="species", y="petal_length",data=iris) 
```

```python
sns.regplot(x="sepal_width",y="sepal_length", data=iris,color='hotpink')#plot data and a linear regression model
```

```python
plot = sns.distplot(iris.sepal_length,color="g")
'''Distplot stands for distribution plot,
it takes as input an array and plots a 
curve corresponding to the distribution of points in the array'''
```
